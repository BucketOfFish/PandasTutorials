```python
%matplotlib inline
import matplotlib
import pandas as pd
pd.set_option('display.mpl_style', 'default')
```

```python
bikes = pd.read_csv('pandas-cookbook/data/bikes.csv', sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
```

```python
bikes['Berri 1'].plot(figsize=(15,5))
```

```python
berri_bikes = bikes[['Berri 1']] # double brackets make this a dataframe
```

```python
berri_bikes.index.weekday # 0 is Monday
```

```python
berri_bikes['weekday'] = berri_bikes.index.weekday
```

```python
berri_bikes[:5]
```

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
weekday_counts.plot(kind='bar', figsize=(15,5))
```

```python
def list_range(my_list):
    return max(my_list) - min(my_list)

berri_bikes.groupby('weekday').aggregate(list_range)
```
